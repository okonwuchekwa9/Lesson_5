```python
# Define file path
fp = 'kumpula-june-2016-w-metadata.txt'
#Read in data from the file
data = pd.read_csv(fp, skiprows=8)
#temp_data = pd.read_csv('kumpula-june-2016-w-metadata.txt', skiprows=8, usecols=['YEARMODA', 'TEMP'])
```

```python
# Define a new column "Diff"
data["DIFF"] = 0.0
#Check what data looks like
data.head()
```

```python
# Check datatypes
data["DIFF"].dtypes
```

```python
# Calculate max min difference
data["DIFF"] = data["MAX"] - data["MIN"]
# Check result
data.head()
```

```python
# creat a new column and convert temp fahrenheit to celsius
data["TEMP_CELCIUS"] = (data["TEMP"]-32) / (9/5)
# check output
data.head()
```

```python
# create a column and convert temp celsius to Kelvin
data["TEMP_KELVIN"] = data["TEMP_CELCIUS"] + 273
#check output
data.head()
```
